```
import numpy as np
from keras.datasets import mnist
from keras.models import Sequential
from keras.utils import np_utils
from keras.layers import Dense,Conv2D,MaxPooling2D,Flatten
from keras.optimizers import Adam

(x_train,y_train),(x_test,y_test) = mnist.load_data()
x_train = x_train.reshape(-1,1,28,28)/255  #normalize 收敛更快
x_test = x_test.reshape(-1,1,28,28)/255
y_train = np_utils.to_categorical(y_train,num_classes=10)
y_test = np_utils.to_categorical(y_test,num_classes=10)
#network structure
#convolutinal layer output 32*28*28
my_model = Sequential()
my_model.add(Conv2D(
    batch_input_shape=(None,1,28,28),
    filters=32,
    kernel_size=(5,5),
    activation='relu',
    padding='same',
    data_format='channels_first'   
))

#pooling output 32*14*14
my_model.add(MaxPooling2D(
    pool_size=2,
    strides=2,
    padding='same',
    data_format='channels_first'  
))

#convolutinal again output 64*14*14
my_model.add(Conv2D(64,(5,5),padding='same',activation='relu',data_format='channels_first'))

#pooling output 64*7*7
my_model.add(MaxPooling2D(2,2,padding='same',data_format='channels_first'))

#fully connected layer 1  output 1024
my_model.add(Flatten())
my_model.add(Dense(1024,activation='relu'))

#fully connect layer 2  output 10
my_model.add(Dense(10,activation='softmax'))

#compile the model
#adam = Adam(lr=1e-4)
my_model.compile(optimizer='adam',
             loss='categorical_crossentropy',
             metrics=['accuracy'])

print('Training-----')
my_model.fit(x_train,y_train,epochs=1,batch_size=128)

print('Testing-----')
loss,accuracy = my_model.evaluate(x_test,y_test)
print('loss:{}  accuracy:{}'.format(loss,accuracy))

```
